# Exercise 1: Create a new column that contains only the zipcode.

```python
dataB = []
with open('../DataYelp/yelp_academic_dataset_business.json') as f:
    for line in f:
        dataB.append(simplejson.loads(line))
```

```python
dfB = pd.DataFrame(dataB[:5000]) # only use the first 5000 to increase speed
dfB.head()
```

```python
def find_zipcode(a_string):
    '''Takes a string containing a zip code and returns the zip code as a string'''
    try:
        zipcode_with_space = re.search(r'\s\d{5}\b', a_string)
        zipcode = re.search(r'\b\d{5}\b', zipcode_with_space.group())
        
        return zipcode.group()
    except:
        return None

    
```

```python
dfB['zip_code'] = dfB['full_address'].apply(find_zipcode)
```

```python
dfB.head()
```

# Exercise 2: The table contains a column called 'categories' and each entry in this column is populated by a list. We are interested in those businesses that restaurants. Create a new column 'Restaurant_type' that contains a description of the restaurant based on the other elements of 'categories. 
## That is, if we have '[Sushi Bars, Japanese, Restaurants]' in categories the 'Restaurant_type will be '{'SushiBars': 1, 'Japanese': 1, 'Mexican': 0, ...}'

```python
RestCat = {}

for row in dfB['categories']:
    if 'Restaurants' in row:
        for item in row:
            if item != 'Restaurants':
                RestCat[item] = 0
        
```

```python
list(RestCat)[:10]
```

```python
def Cat(row):
    d = dict(RestCat)
    if 'Restaurants' in row:
        for item in row:
            if item != 'Restaurants':
                d[item] = 1
            else: 
                d[item] = 0
    else:
        d = 'Remove'

    return d
```

```python
dfB['restaurant_type'] = dfB['categories'].apply(Cat)
```

# Exercise 3: Lets clean the 'attributes' column. The entries in this column are dictionaries. We need to do two things: 
## 1) Turn all the True or False in the dictionary to 1s and 0s.
## 2) There are some entries within dictionaries that are dictionaries themselves, lets turn the whole entry into just one dictionary, for example if we have 
### '{'Accepts Credit Cards': True, 'Alcohol': 'none','Ambience': {'casual': False,'classy': False}}' 
### then turn it into
### '{'Accepts Credit Cards':1, 'Alcohol_none': 1, 'Ambience_casual': 0, 'Ambience_classy': 0}'. 
### There might be other entries like {'Price Range': 1} where the values are numerical so we might want to change that into {'Price_Range_1': 1}.

*The reason we modify categorical variables like this is that machine learning algorithms cannot interpret textual data like "True" and "False". They need numerical inputs such as 1 and 0.*

```python
def Flat(row):
    d = row['attributes']
    d1 = {}
    
    for el in d.keys():
        if type(d[el]) == bool:
            d1[el] = int(d[el])
            
        elif type(d[el]) == str:
            label = '{}_{}'.format(el, d[el])
            d1[label] = 1
            
        elif type(d[el]) == int:
            label = '{}_{}'.format(el, d[el])
            d1[label] = 1
            
        elif type(d[el]) == dict:
            for elem in d[el].keys():
                label = '{}_{}'.format(el, elem)
                d1[label] = int(d[el][elem])
                
        else:
            print(el, d[el])
            
    return d1
```

```python
dfB['attributes_cleaned'] = dfB.apply(lambda x: Flat(x), 1)
```

# Exercise 4: Create a new column for every day of the week and fill it with the amount of hours the business is open that day.

*Your approach should handle businesses that stay open late like bars and nightclubs.*

```python
dfB['hours'][0]
```

```python
days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
```

```python
FMT = '%H:%M'

def Time(row, date):
   
    try:
        closing = datetime.strptime(row['hours'][date]['close'], FMT)
        opening = datetime.strptime(row['hours'][date]['open'], FMT)
        time = closing - opening

        if time < timedelta(days=0): # checks it the close time is actually on the next day (i.e. 2:00 am)
            time += timedelta(days=1) # adds 1 day of time if close is after midnight 
    except:
        time = timedelta(days=0)

    return time
```

```python
for day in days:
    dfB[day] = dfB.apply(lambda x: Time(x, day), 1)
```

# Exercise 5: Create a table with the average review for a business.

*You will need to pull in a new json file and merge DataFrames for the next 2 exercises.*

```python
dataR = []
with open('../DataYelp/yelp_academic_dataset_review.json') as f:
    for line in f:
        dataR.append(simplejson.loads(line))
```

```python
dfR = pd.DataFrame(dataR[:5000])
dfR.head()
```

```python
avgR = dfR.groupby('business_id')['stars'].mean().reset_index()
avgR.columns = ['business_id', 'star_avg']
avgR.head()
```

```python
dfR = dfR.merge(avgR, how='left', on='business_id')
dfR.head()
```
